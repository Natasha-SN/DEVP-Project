```python
# if a pokemon can be regarded as legendary or not 
_=sns.countplot(x=df['Legendary'].value_counts(), data=df)
```

```python
# relationship between special attack and special defence of a pokemon
_=sns.relplot(x ="Sp. Atk", y ="Sp. Def",
            data = df)
```

```python
# which generation a particular pokemon belongs to
_=sns.countplot(x='Generation', data=df)
```

```python
#Comparing attack and defense capabilities of Fire and water type pokemons 

_=sns.relplot(x="Attack", y="Defense", hue= "Type 1",  data=df )
```

```python
#Comparing attack and defense capabilities of Legendary pokemons with Non-legendary ones
_=sns.relplot(x="Attack", y="Defense", hue= "Legendary",  data=df)
```

```python
# speed comaprison over various generations of Pokemons

_=sns.lineplot(data=df, x="Generation", y="Speed", estimator = np.mean)
```
